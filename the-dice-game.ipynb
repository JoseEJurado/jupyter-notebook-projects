# The Dice Game
The game consists of 3 components: a dice, a pile of chips, and the players.

## Rules of the Game
The rules are simple.  
There is a pile of 20 chips.  
Each person takes turns rolling a dice.  
Their roll indicates how many chips they remove from the stack.  
The person to remove the last chip from the pile loses.  

```python
# Imports
import random
import time

# Dice mechanic

def dice_roll():
    r = random.randint(1,6)
    return r

# Game Explanation

print(f"""Welcome to \033[1mThe Dice Game\033[0m.\n
This game requires:\n- 2 players\n- 1 Die\n- A pile of chips.\n
The rules are simple.\nEach player takes turns rolling the dice.
The number rolled is the amount removed from the chip pile.
The player to remove the last chip loses.
The default amount of chips in a pile is 20. This can be changed.""")

# Number of chips

chip_stack = 20
chip_change = str(input("\nWould you like to change the amount of starting chips?\n[Y / N]\n").lower())

if chip_change == 'y':
    try:
        chip_stack = input("\nHow many chips would you like to start with?\n")
        chip_stack = int(chip_stack)
    except ValueError:
        print("Please note a valid amount.")
else:
    pass

print(f"\nThe chip pile now consists of {chip_stack} chips.".format())

# Player names

print(f"\nNow, let's get acquainted with the players.\n")

p1 = str(input("Player 1, what is your name?\n").capitalize())
p2 = str(input("Player 2, what is your name?\n").capitalize())

print(f"\nThank you for playing, {p1} and {p2}. We will now determine who goes first.\n\
Each of you will roll a die and the higher number will go first.".format())

# Player order

print(f"\nRolling for {p1}...\n")
for i in range(3):
    print("Rolling...")
    time.sleep(1.5)
p1_roll = dice_roll()
print(f"\n{p1}'s roll: {p1_roll}")

print(f"\nRolling for {p2}...\n")
for i in range(3):
    print("Rolling...")
    time.sleep(1.5)
p2_roll = dice_roll()
print(f"\n{p2}'s roll: {p2_roll}")

# Failsafe

while p1_roll == p2_roll:
    print(f"\nHuh, a tie. Let's roll again.")
    p1_roll = dice_roll()
    p2_roll = dice_roll()
    print(f"\n{p1} you rolled a: {p1_roll}. {p2}, you rolled a: {p2_roll}.")
    
if p1_roll > p2_roll:
    print(f"\nPlayer {p1} goes first.")
    p_first = p1
    p_second = p2
    
else:
    print(f"\nPlayer {p2} goes first.")
    p_first = p2
    p_second = p1
    
# Game start

moves = 0
first_move = True
game = True

while game:
    while chip_stack != 0 or chip_stack <= 0:
        while first_move is True:
            key_press = input(f"\n{p_first}, press Enter to roll.\n")
            roll = dice_roll()
            time.sleep(1)
            print(f"\n{p_first}, you rolled a: {roll}.\n")
            chip_stack -= roll
            time.sleep(1)
            print(f"\nThere are now {chip_stack} chips left.\n")
            if chip_stack == 0 or chip_stack < 0:
                print(f"\n{p_first}, you have removed the last chip. You lose.\n")
                game = False
                first_move = None
            else:
                first_move = False
        while first_move is False:
            key_press = input(f"\n{p_second}, press Enter to roll.\n")
            roll = dice_roll()
            time.sleep(1)
            print(f"\n{p_second}, you rolled a: {roll}.\n")
            chip_stack -= roll
            time.sleep(1)
            print(f"\nThere are now {chip_stack} chips left.\n")
            if chip_stack == 0 or chip_stack < 0:
                print(f"\n{p_second}, you have removed the last chip. You lose.\n")
                game = False
                first_move = None
            else:
                first_move = True
```
