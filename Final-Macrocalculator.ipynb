## Calorie and Macronutrient Calculator Code
### Step 1:
Using inputed weight, height, age, and sex the calculator first calculates the users BMR. 
### Step 2:
Then using the BMR and chosen activity level the TDEE is calculated. 
### Step 3:
The users chosen weight goal and the TDEE is used to calculate their suggested calories.
### Step 4: 
Finally, a 40/30/30 macro split is applied to the calculated calories.

```python
#User Information Function
def userinfo():
    """Obtains the user's information and calculates bmr."""
    weight = str(input("Please enter your weight (in lbs): "))
    while not weight.isdigit():
        print("Please enter a number. ")
        weight = str(input("Please enter your weight (in lbs): "))
    weight = float(weight)
    
    height = str(input("Please enter your height (in inches): "))
    while not height.isdigit():
        print("Please enter a number. ")
        height = str(input("Please enter your height (in inches): "))
    height = float(height)
        
    age = int(input("Please enter your age(18-80): "))
    while age not in range(18,80):
        print("Please enter a valid age. ")
        age = int(input("Please enter your age(18-80): "))
    age = float(age)      
            
    sex = input("Female or Male: ")
    while sex not in ['Female', 'female', 'Male', 'male']:
        print("Sorry, please choose one of the options. ")
        sex = input("Female or Male: ")
    
    if sex in ['Female', 'female']:
        bmr = 655 + (4.3 * weight) + (4.7 * height) - (4.7 * age)
        return bmr
    elif sex in ['Male', 'male']:
        bmr = 66 + (6.3 * weight) + (12.9 * height) - (6.8 * age)
        return bmr

#User Activity Levels Function
def useract(bmr):
    """Calculates the user's total daily energy expenditure (TDEE)."""
    activity = input("""What is your activity level?
    1) Sedentary (little to no exercise)
    2) Lightly Active (light exercise/sports 1-3 days/week
    3) Moderately Active (moderate exercise/sports 3-5 days/week
    4) Very Active (hard exercise/sports 6-7 days/week
    5) Extra Active (very hard exercise/sports AND physical job/2x training)
    Simply type in the corresponding number. """)
    while activity not in ['1', '2', '3', '4', '5']:
        print("Sorry, that is not a viable option. Please try again. ")
        activity = input("""What is your activity level?
    1) Sedentary (little to no exercise)
    2) Lightly Active (light exercise/sports 1-3 days/week
    3) Moderately Active (moderate exercise/sports 3-5 days/week
    4) Very Active (hard exercise/sports 6-7 days/week
    5) Extra Active (very hard exercise/sports AND physical job/2x training)
    Simply type in the corresponding number. """)
    
    if activity == '1':
        tdee = bmr * 1.2
    elif activity == '2':
        tdee = bmr * 1.375
    elif activity == '3':
        tdee = bmr * 1.55
    elif activity == '4':
        tdee = bmr * 1.725
    elif activity == '5':
        tdee = bmr * 1.9
    return (tdee)

#User Goal Function
def usergoal(tdee):
    """Calculates the user's caloric needs based on their fitness goals."""
    goal = input("""What are your goals?
    1) Lose Weight
    2) Maintain Weight
    3) Gain Weight
    Simply type in the corresponding number. """)
    while goal not in ['1', '2', '3']:
        print("Sorry, that is not a viable option. Please try again. ")
        goal = input("""What are your goals?
    1) Lose Weight
    2) Maintain Weight
    3) Gain Weight
    Simply type in the corresponding number. """)
        
    if goal == '1':
        goal_tdee = tdee - 500
        print("In order for you to lose weight at a rate of 1 lb/week, you have to eat {} calories/day.".format(goal_tdee))
        return(goal_tdee)
    elif goal == '2':
        goal_tdee = tdee
        print("In order for you to maintain your weight, you have to eat {} calories/day.".format(goal_tdee))
        return(goal_tdee)
    elif goal == '3':
        goal_tdee = tdee + 500
        print("In order for you to gain weight at a rate of 1 lb/week, you have to eat {} calories/day.".format(goal_tdee))
        return(goal_tdee)

#User Carb, Fat, Protein Intake
def userCFP(goal_tdee):
    """Provides the user with a 40/30/30 macro split."""
    carbscal = goal_tdee * 0.4
    fatscal = goal_tdee * 0.3
    proteincal = goal_tdee * 0.3
    carbs = carbscal / 4
    fats = fatscal / 9
    proteins = proteincal / 4
    print("""For optimal results, your macros are split 40/30/30 (40% carbs, 30% fats, 30% proteins). 
    You must eat a total of {:6.1f}g of carbs, {:6.1f}g of fats, and {:6.1f}g of proteins.""".format(carbs, fats, proteins))
    print('Thanks for using the macrocalculator! Best of luck reaching your goals!')

userCFP(usergoal(useract(userinfo())))
```

## Webscraping code
### Step 1:
Ask user if they are interested in help finding macronutrients sources from Walmart.
### Step 2:
If not interested, wish them good luck, but if interested ask which macronutrient they are interested in searching.
### Step 3:
Web scrape Walmart's website for related items and provide a list of items and prices.

```python
def scrape(url):
                """Returns parsed HTML content from desired url,
                which specifies a store and product category"""
                response = urlopen(url)
                content = response.read()
                soup = BeautifulSoup(content, 'html.parser')
                return soup
            
def get_items(soup):
    """Return a list of item names"""
    item_list = soup.find_all('a', attrs = {'class': 'product-title-link line-clamp line-clamp-2 truncate-title'})
    items = []
    for item in item_list:
        items.append(item.get_text())
    return items
                
def get_prices(soup):
    """Returns a list of prices"""
    price_list = soup.select('span.price > span.visuallyhidden')
    prices = []
    for price in price_list:
        prices.append(price.get_text())
    return prices

def usershop():
    """Helps the user find cheap products."""
    ask = input("""Would you like some help looking for grocery items at Walmart?
    Yes / No: """)
    while ask not in ['no', 'No', 'yes', 'Yes']:
        print("Sorry, that's not an available option. Please try again.")
        ask = input("""Would you like some help looking for cheap grocery items at Walmart?
        Yes / No: """)
    if ask in ['no', 'No']:
        print('')
    if ask in ['yes', 'Yes']:
        # Macronutrient User Option
        macroask = input("""What source of nutrient are you interested in searching?
        1. Carbohydrates
        2. Fats
        3. Proteins
        Simply type in the corresponding number. """)
        while macroask not in ['1', '2', '3']:
            print("Sorry, please choose one of the previous options.")
            macroask = input("""What source of nutrient are you interested in?
            1. Carbohydrates
            2. Fats
            3. Proteins
            Simply type in the corresponding number. """)

        #(C)arbohydrates, (W)almart
        if macroask == '1':
            c_walmart = scrape('https://www.walmart.com/browse/meal-solutions-grains-pasta/grains-rice/976759_976794_1001474')
            cw_items = get_items(c_walmart)
            cw_prices = get_prices(c_walmart)
            print("Here is a list of some of Walmart's available carb sources: ")
            print(cw_items)
            print('')
            print('And a list of their corresponding prices: ')
            print(cw_prices)
                
        #(F)ats, (W)almart
        elif macroask == '2':
            f_walmart = scrape('https://www.walmart.com/browse/food/fresh-food/976759_1071964?cat_id=976759_1071964_976788')
            fw_items = get_items(f_walmart)
            fw_prices = get_prices(f_walmart)
            print("Here is a list of some of Walmart's available fat sources: ")
            print(fw_items)
            print('')
            print('And a list of their corresponding prices: ')
            print(fw_prices)   
                
        #(P)rotein, (W)almart
        elif macroask == '3':
            p_walmart = scrape('https://www.walmart.com/browse/food/fresh-food/976759_1071964?cat_id=976759_1071964_976796')
            pw_items = get_items(p_walmart)
            pw_prices = get_prices(p_walmart)
            print("Here is a list of some of Walmart's available protein sources: ")
            print(pw_items)
            print('')
            print('And a list of their corresponding prices: ')
            print(pw_prices)  
                
    print("Thank you for using the food search feature! Happy dieting! ")
        
usershop()
```
